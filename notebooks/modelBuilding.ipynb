```python
#data X and Y walata beda ganna ona
x = data.drop('selling_price', axis=1)
y = data['selling_price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape, x_test.shape
```

```python
# Train krnna Algorithm ekak ganna ona
```

```python
def model_acc(model):
    model.fit(x_train,y_train)
    acc = model.score(x_test, y_test)
    print(str(model) + '-->' + str(acc))
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
model_acc(LR)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)
```

```python
from sklearn.tree import DecisionTreeRegressor
Dt = DecisionTreeRegressor()
model_acc(Dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
model_acc(RF)
```

```python
#Train karann RandomForestRegressor ek gamu
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10,50,100],
              'criterion':['squared_error','abolute_error','poisson']}

grid_obj = GridSearchCV(estimator = RF , param_grid = parameters)

grid_fit = grid_obj.fit(x_train, y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test, y_test)
# hypoparameter tune karata passe 88%k accuracy ekak enw 
```

```python
# Model eka save kara gamu
```

```python
import pickle
with open('PricePredictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
